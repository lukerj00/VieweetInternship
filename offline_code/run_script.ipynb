```python
#from azureml.core import Experiment, ScriptRunConfig
#from azureml.widgets import RunDetails
#from azureml.tensorboard import Tensorboard

#Set the script parameters
script_params = [
    '--data_dir',   str(data_ref),
    '--model_dir',  str(model_data_ref),    
    '--log_dir',    './logs', 
    '--classifier', 'full',
    '--batch_size', 1,
    '--dummy',
    '--incl_metadata'
]

src = ScriptRunConfig(source_directory='.',
                      arguments=script_params,
                      compute_target = compute_target,
                      environment = env,
                      script='test_multi.py')
src.run_config.data_references = {data_ref.data_reference_name: data_ref.to_config(),
                                  model_data_ref.data_reference_name: model_data_ref.to_config()} 

experiment = Experiment(workspace = ws, name ='rc_test')
run = experiment.submit(config=src, 
                        tags={'classifier':'full',
                              'training context':'unbalanced, unweighted, augmented',
                              'codex':'full_wmeta',
                              'model':'ResNet10',
                              'image size':'64x128',
                              'dataset':'full'                                                                                             ,
                        })

RunDetails(run).show()
run.wait_for_completion()
```
